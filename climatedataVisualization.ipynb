```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('/content/sample_data/Climate_full.csv')
```

```python
print(dataset)
```

```python
x=range(1,24)
```

```python
df1=dataset.iloc[1:24,3:4]

df2=dataset.iloc[1:24,4:5]
df3=dataset.iloc[1:24,5:6]

y1=df1['MaxTemp'].tolist()
y2=df2['MinTemp'].tolist()
y3=df3['TempC'].tolist()
```

```python
plt.figure(figsize=(30,10))
plt.plot(x,y1,label='Max Temperature')
plt.plot(x,y2,label='Min Temperature')
plt.plot(x,y3,label='TempC')
plt.title('weather data')
plt.xlabel('Year')
plt.legend(loc='upper right')
plt.show()
```
